```python
import pandas as pd
import numpy as np
```

```python
url=('/content/Marketing campaign dataset.csv')
```

```python
df=pd.read_csv(url)
print(df.info())
```

```python
print(df.head())
```

```python
df.fillna(0, inplace=True)
```

```python
df['CTR_%'] = np.where(df['impressions'] > 0, (df['clicks'] / df['impressions']) * 100, 0)

df['CPC_USD'] = np.where(df['clicks'] > 0, df['media_cost_usd'] / df['clicks'], 0)

df['CPM_USD'] = np.where(df['impressions'] > 0, (df['media_cost_usd'] / df['impressions']) * 1000, 0)

print(df[['channel_name', 'CTR_%', 'CPC_USD', 'CPM_USD']].head())
```

```python
df.to_csv('cleaned_marketing_data.csv', index=False)
```
